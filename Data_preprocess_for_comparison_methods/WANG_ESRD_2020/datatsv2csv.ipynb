```python
#数据划分
import pandas as pd
pd.set_option('display.max_columns', None)
# 读取TSV文件并将其转换为DataFrame
genera = pd.read_csv('genera.tsv', sep='\t')
metadata = pd.read_csv('metadata.tsv', sep='\t')
mtb = pd.read_csv('mtb.tsv', sep='\t')
mtb_map = pd.read_csv('mtb.map.tsv', sep='\t')
# 显示DataFrame|
mtb
```

```python
numeric_cols = mtb.select_dtypes(include='number')
# 计算每行的和
row_sums = numeric_cols.sum(axis=1)
print(row_sums)
```

```python
numeric_cols = genera.select_dtypes(include='number')
# 计算每行的和
row_sums = numeric_cols.sum(axis=1)
print(row_sums)
```

```python
from sklearn.model_selection import train_test_split


# 划分微生物表的测试集
microbe_positive = genera[genera['Sample'].str.startswith('CON')]
microbe_negative = genera[genera['Sample'].str.startswith('KD')]

microbe_positive_train, microbe_positive_test = train_test_split(microbe_positive, test_size=0.2, random_state=42)
microbe_negative_train, microbe_negative_test = train_test_split(microbe_negative, test_size=0.2, random_state=42)

# 合并正负样本的测试集
microbe_test = pd.concat([microbe_positive_test, microbe_negative_test])
microbe_train = pd.concat([microbe_positive_train, microbe_negative_train])

# 选择对应的代谢物作为测试集
metabolite_test = mtb[mtb['Sample'].isin(microbe_test['Sample'])].copy()
metabolite_train = mtb[~mtb['Sample'].isin(microbe_test['Sample'])].copy()


#选择对应的样本标签
metadata_test = metadata[metadata['Subject'].isin(microbe_test['Sample'])].copy()
metadata_train = metadata[~metadata['Subject'].isin(microbe_test['Sample'])].copy()
# 在测试集样本名称前加上"Validation."
metadata_test['Subject'] = 'Validation.' + metadata_test['Subject']
# 在测试集样本名称前加上"Validation."
microbe_test['Sample'] = 'Validation.' + microbe_test['Sample']
metabolite_test['Sample'] = 'Validation.' + metabolite_test['Sample']

metadata_all = pd.concat([metadata_test, metadata_train])
microbe_all = pd.concat([microbe_train, microbe_test])
metabolite_all = pd.concat([metabolite_train, metabolite_test])
```

```python
metabolite_all= metabolite_all.sort_index()
metabolite_all
```

```python
microbe_all = microbe_all.sort_index()
microbe_all
```

```python
# 导出为CSV格式
microbe_all.to_csv('genera.csv', index=False)
metabolite_all.to_csv('mtb.csv', index=False)
```
