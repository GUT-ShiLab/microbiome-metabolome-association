```python
import pandas as pd
pd.set_option('display.max_columns', None)
# TSVDataFrame
genera = pd.read_csv('genera_filtered.csv')
metadata = pd.read_csv('metadata.tsv', sep='\t')
mtb = pd.read_csv('mtb_filtered.csv')
mtb_map = pd.read_csv('mtb.map.tsv', sep='\t')
# DataFrame
mtb
```

```python
samples = mtb['Sample']
```

```python
import numpy as np

# 
mtb = mtb.drop(columns=['Sample'])  # 'Sample'


# 
for index, row in mtb.iterrows():
    min_nonzero_value = row[row != 0].min()
    min_nonzero_half = min_nonzero_value / 2
    mtb.loc[index] = row.replace(0, min_nonzero_half)

mtb.insert(0, 'Sample', samples)
mtb
```

```python
import pandas as pd

# dfDataFrame
# read_csv
# df = pd.read_csv('your_data.csv')

# 
samples = mtb['Sample']

# DataFrame
mtb.drop(columns=['Sample'], inplace=True)

# 
def normalize_row(row):
    # 
    row_sum = row.sum()
    # 
    return row / row_sum if row_sum != 0 else row

# DataFrame
normalized_mtb = mtb.apply(normalize_row, axis=1)

# DataFrame
normalized_mtb.insert(0, 'Sample', samples)

# DataFrame
normalized_mtb
```

```python
normalized_mtb.to_csv('normalized_mtb.csv', index=False)
```

```python
normalized_mtb = normalized_mtb.set_index(normalized_mtb.columns[0])  # 
genera = genera.set_index(genera.columns[0])  # 
normalized_mtb
```

```python
# 
training_genera = genera[~genera.index.str.contains('Validation')]
testing_genera = genera[genera.index.str.contains('Validation')]

training_normalized_mtb = normalized_mtb[~normalized_mtb.index.str.contains('Validation')]
testing_normalized_mtb = normalized_mtb[normalized_mtb.index.str.contains('Validation')]

# CSV
training_genera.to_csv('ENVIM_data1/training.mic.data.csv', index=True)
testing_genera.to_csv('ENVIM_data1/testing.mic.data.csv', index=True)

training_normalized_mtb.to_csv('ENVIM_data1/training.mtb.data.csv', index=True)
testing_normalized_mtb.to_csv('ENVIM_data1/testing.mtb.data.csv', index=True)
```
